```python
import matplotlib.pyplot as plt

# Extract the data from your results
dataset_sizes = [1000, 10000, 100000, 250000, 500000, 1000000]

# Execution times
insertion_times = [0.016, 1.72, 198.96, 1306.05, 5666.80, 27513.09]
merge_times = [0.001, 0.02, 0.23, 0.62, 1.38, 2.97]

# Create the graph
plt.figure(figsize=(12, 8))
plt.plot(dataset_sizes, insertion_times, 'ro-', label='Insertion Sort', linewidth=2, markersize=8)
plt.plot(dataset_sizes, merge_times, 'bo-', label='Merge Sort', linewidth=2, markersize=8)

plt.xlabel('Dataset Size (number of elements)')
plt.ylabel('Execution Time (seconds)')
plt.title('Insertion Sort vs Merge Sort Performance Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
